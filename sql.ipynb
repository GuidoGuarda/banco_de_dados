# WHERE
## Operadores    -     Descrição
    =                   Igual
    >                   Maior que
    <                   Menor que
    >=                  Maior ou igual que
    <=                  Menor ou igual que
    <>                  Diferente de
    AND                 Operador Lógico E
    OR                  Operador Lógico OU   
## selecionando dados específicos 
SELECT * FROM chart
WHERE name = 'João Ferreira'
## selecionando produtos com cores diferentes
SELECT * FROM production.products
WHERE color = 'blue' or color = 'red'
## selecionando produtos especificando valores
SELECT * FROM production.products
WHERE price > 1500 and price <= 5000
## selecionando produtos retirando um tipo excluindo um tipo
SELECT * FROM production.products
WHERE color <> 'black'
## selecionando produtos com peso entre 600 e 800 kg
